```python
import pandas as pd 
```

```python
 productionunit_df = pd.read_excel("Data for Group analytic report-1.xlsx", sheet_name="Production plants Data") 
sales_df = pd.read_excel("Data for Group analytic report-1.xlsx", sheet_name="Sales data (annual)") 
logistics_df = pd.read_excel("Data for Group analytic report-1.xlsx", sheet_name="Logistics data") 
products_df = pd.read_excel("Data for Group analytic report-1.xlsx", sheet_name="Parts and Raw material sourcing") 
```

```python
# Example for Sales
key_vars = ['Type A Units sold per year', 'Type B Units sold per year', 
            'Type C Units sold per year', 'Revenue (£m)', 'Total Number of Returns']

summary = sales_df[key_vars].describe().T  # mean, std, min, max
summary['range'] = summary['max'] - summary['min']
print(summary)
```

```python
def full_summary(df, name):
    print(f"\n--- {name} Summary ---")
    print(df.describe())
```

```python
# Run summaries
full_summary(productionunit_df, "Manufacturing")
full_summary(sales_df.drop(columns='Country'), "Sales")
full_summary(logistics_df, "Logistics")
full_summary(products_df.drop(columns=['Product type Id','Product type']), "Products")
```

```python
if 'Countries' in logistics_df.columns:
    logistics_df = logistics_df.rename(columns={'Countries': 'Country'})

# Proceed with merge only if 'Country' exists in all
if all('Country' in df.columns for df in [productionunit_df, logistics_df, sales_df]):
    merged_df = pd.merge(productionunit_df, logistics_df, on='Country', how='left')
    merged_df = pd.merge(merged_df, sales_df, on='Country', how='left')
    print("\nMerged Data:")
    print(merged_df.head())
else:
    print("❌ 'Country' column not found in one or more DataFrames")
```

### Correlation Heatmaps

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales_by_type = sales_df[['Type A Units sold per year', 'Type B Units sold per year', 'Type C Units sold per year']]
sales_by_type.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title('Sales by Product Type')
plt.ylabel('Units Sold')
plt.xlabel('Index')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=sales_df, x='Advertisement expenditure (£)', y='Revenue (£m)', hue='Country')
plt.title("Revenue vs Advertisement Spend")
plt.xlabel("Ad Spend (£)")
plt.ylabel("Revenue (£m)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=sales_df, x='Country', y='Total Number of Returns', palette='Reds_r')
plt.title("Total Product Returns by Country")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=sales_df, y='Average yearly Fuel price £ (per L)')
plt.title("Fuel Price Distribution")
plt.ylabel("Fuel Price (£)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=logistics_df, x='Number of distribution centres (sale points)', 
                y='Logistic costs (£)', size='Customer satisfaction rate (5*)', hue='Country', sizes=(20, 200))
plt.title("Distribution Centres vs Logistic Cost")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(data=logistics_df, x='Average shipping time (from assembly plants to selling point)', bins=8)
plt.title("Distribution of Shipping Times")
plt.xlabel("Days")
plt.tight_layout()
plt.show()
```

```python
top_revenue = sales_df.sort_values(by='Revenue (£m)', ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(data=top_revenue, x='Country', y='Revenue (£m)', palette='Blues_d')
plt.title("Top Countries by Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(sales_df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix - Sales Data")
plt.tight_layout()
plt.show()
```
